
# 📊 Análise Exploratória e Série Temporal — Netflix Dataset

Este notebook realiza:
- **Exercício 1:** Análise exploratória de uma variável numérica (*release_year*)
- **Exercício 2:** Análise de série temporal (quantidade de títulos por ano)

Visualizações com **Plotly** e **tema customizado**.


```python

import pandas as pd
import plotly.express as px
import plotly.io as pio
```

```python
!wget https://raw.githubusercontent.com/daninagem/tccpuc/main/notebook/netflix_titles.csv
```

```python

# 🎨 Tema customizado
pio.templates["custom"] = dict(
    layout=dict(
        font=dict(family="Arial", size=14, color="#2D3436"),
        title=dict(font=dict(size=20, color="#2D3436")),
        plot_bgcolor="#FFFFFF",
        paper_bgcolor="#F5F6FA",
        xaxis=dict(showgrid=True, gridcolor="#EAEAEA"),
        yaxis=dict(showgrid=True, gridcolor="#EAEAEA")
    )
)

pio.templates.default = "custom"
```

```python

# 📥 Carregar dataset
# ⚠️ Coloque o arquivo netflix_titles.csv na mesma pasta do notebook
df = pd.read_csv("netflix_titles.csv")
df = df.dropna(subset=["release_year"])

df.head()
```


## 🧪 Exercício 1 — Análise Exploratória

**Variável escolhida:** `release_year`  
Representa o ano de lançamento dos títulos na Netflix.


```python

fig = px.histogram(
    df,
    x="release_year",
    title="Distribuição dos anos de lançamento",
    color_discrete_sequence=["#6C5CE7"]
)

fig.update_layout(title_x=0.5, bargap=0.1)
fig.show()
```

```python

fig = px.box(
    df,
    y="release_year",
    title="Boxplot dos anos de lançamento",
    color_discrete_sequence=["#00CEC9"]
)

fig.update_layout(title_x=0.5)
fig.show()
```

```python

print("Média:", df["release_year"].mean())
print("Mediana:", df["release_year"].median())
print("Desvio padrão:", df["release_year"].std())
print("Q1:", df["release_year"].quantile(0.25))
print("Q3:", df["release_year"].quantile(0.75))
```


### 📊 Interpretação

A distribuição apresenta maior concentração após os anos 2000, indicando crescimento recente de conteúdos.

O boxplot mostra poucos valores antigos como possíveis outliers.

A média e mediana próximas indicam distribuição relativamente equilibrada.



## ⏱️ Exercício 2 — Série Temporal


```python

df_time = df.groupby("release_year").size().reset_index(name="quantidade")
df_time.head()
```

```python

fig = px.line(
    df_time,
    x="release_year",
    y="quantidade",
    title="Quantidade de lançamentos por ano"
)

fig.update_traces(line=dict(color="#6C5CE7", width=3))
fig.update_layout(title_x=0.5)
fig.show()
```

```python

df_time["media_movel"] = df_time["quantidade"].rolling(window=5).mean()

fig = px.line(
    df_time,
    x="release_year",
    y="media_movel",
    title="Média móvel de lançamentos"
)

fig.update_traces(line=dict(color="#00CEC9", width=3))
fig.update_layout(title_x=0.5)
fig.show()
```


### 📈 Interpretação

Há uma tendência clara de crescimento ao longo do tempo.

A média móvel suaviza variações e facilita a visualização do padrão geral.

Limitação: pode esconder variações importantes em períodos específicos.



## ✅ Conclusão

A análise exploratória e temporal permitiu identificar padrões relevantes nos dados, evidenciando crescimento ao longo do tempo e distribuição concentrada em períodos recentes.

O uso de visualizações foi essencial para compreender melhor os dados e extrair insights.
