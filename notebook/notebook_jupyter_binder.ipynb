## 📥 Carregamento de Dados

````python
import pandas as pd

url = 'https://raw.githubusercontent.com/daninagem/tccpuc/main/dataset/pnadc_ocupados_2015_2025_base.parquet'
df = pd.read_parquet(url)
df.head()
````

---

## H1 — Educação e renda
<a id='H1'></a>

````python
import matplotlib.pyplot as plt

plt.figure()
df.groupby('educacao')['renda'].mean().plot()
plt.title('Educação vs Renda')
plt.grid()
plt.show()
````
